# 2.	Encode categorical variables using one-hot encoding or label encoding. and 3.	If needed you can apply more preprocessing methods on the given dataset.

```python
# numerical variables
df_train_num = df_train[['Age','SibSp','Parch','Fare']]
df_test_num = df_test[['Age','SibSp','Parch','Fare']]

df_train_num = df_train_num.apply(pd.to_numeric, errors='coerce')
df_test_num = df_test_num.apply(pd.to_numeric, errors='coerce')

df_train_num.fillna(df_train_num.mean(), inplace=True)
df_test_num.fillna(df_test_num.mean(), inplace=True)

df_train_num.head()
```

```python
df_test_num.head()
```

```python
# categorical variables
df_train_cat = df_train[['Pclass','Sex','Embarked']]
df_test_cat = df_test[['Pclass','Sex','Embarked']]

df_train_cat.head()
```

```python
# categorical variables
df_test_cat.head()
```

```python
df_train_remaining = df_train[['Survived']]
df_train_remaining.head()
```

```python
# Standardization

from sklearn.preprocessing import StandardScaler
SS = StandardScaler()

#df_train_num = SS.fit_transfor(df_train_num)
#df_train_num = pd.DataFrame(df_train_num)
#df_train_num.columns = ['Age','SibSp','Parch','Fare']
#df_train_num.head()

# Fit and transform the scaler on the training data (only)
df_train_num_scaled = SS.fit_transform(df_train_num)

# Convert the scaled training data back to a DataFrame
df_train_num_scaled = pd.DataFrame(df_train_num_scaled, columns = df_train_num.columns)

# Transform the test data using the same scaler (standardize)
df_test_num_scaled = SS.transform(df_test_num)

# Convert the scaled test data back to a DataFrame
df_test_num_scaled = pd.DataFrame(df_test_num_scaled, columns = df_test_num.columns)

df_train_num_scaled.head()
```

```python
df_test_num_scaled.head()
```

```python
# Label Encoding

from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()

# Concatenate train and test categorical columns
#df_combined_cat = pd.concat([df_train_cat, df_test_cat], axis = 0)

# Apply Label Encoding on the training data
for i in df_train_cat:
    #df_combined_cat[col] = LE.fit_transform(df_combined_cat[col])
    LE.fit(df_train_cat[i])
    df_train_cat[i] = LE.transform(df_train_cat[i])

# Apply Label Encoding to test data
for i in df_test_cat:
    LE.fit(df_train_cat[i])
    df_test_cat[i] = df_test_cat[i].apply(lambda x : LE.transform([x])[0] if x in LE.classes_ else -1)
       
# Split back into train and test sets
#df_train_cat_encoded = df_combined_cat.iloc[:len(df_train_cat), :]
#df_test_cat_encoded = df_combined_cat.iloc[:len(df_test_cat):, :]

df_train_cat.head()
```

# XGBoost

```python
# pip install xgboost
```

- Since our test dataset does not have the actual target label (y_test), we won't be able to directly evaluate the model on the test set using standard performance metrics like accuracy,precision, recall, F1 - score etc., because these metrics require both the predicted values and the actual target values(y_test).
- Since we can only have X_train and y_train, we can evaluate our model using cross-validation on the training data.This technique allows us to assess the performance of our model by splitting our training data into multiple folds. In each fold, part of the data is used for training , and the remaining part is used for validation

```python
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier
xgb = XGBClassifier(use_label_encoder = False, eval_metric = 'logloss')

# performing 5-fold croos-validation on the training data
cv_accuracy = cross_val_score(xgb, X_train, y_train, cv = 5, scoring = 'accuracy')
cv_precision = cross_val_score(xgb, X_train, y_train, cv = 5, scoring = 'precision')
cv_recall = cross_val_score(xgb, X_train, y_train, cv = 5, scoring = 'recall')
cv_f1 = cross_val_score(xgb, X_train, y_train, cv = 5, scoring = 'f1')

# printing the average cross-validation scores
print(f'Cross-validated Accuracy: {cv_accuracy.mean():.4f}')
print(f'Cross-validated Precision: {cv_precision.mean():.4f}')
print(f'Cross-validated Recall: {cv_recall.mean():.4f}')
print(f'Cross-validated F1: {cv_f1.mean():.4f}')
```

```python
# train_test_split

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# split the original training set into a smaller training set and validation set
X_train_split, X_valid, y_train_split, y_valid = train_test_split(X_train, y_train, test_size = 0.2, random_state = 42)

# initialize and train the XGBoost Classifier on the smaller training set
xgb = XGBClassifier(use_label_encoder = False, eval_metric = 'logloss')
xgb.fit(X_train_split, y_train_split)
```

```python
y_valid_pred = xgb.predict(X_valid)
```

```python
# Calculating the evaluation metrics on the validation set

xgb_accuracy = accuracy_score(y_valid, y_valid_pred)
xgb_precision = precision_score(y_valid, y_valid_pred)
xgb_recall = recall_score(y_valid, y_valid_pred)
xgb_f1 = f1_score(y_valid, y_valid_pred)

print('XGBoost Training Accuracy:', xgb_accuracy)
print('XGBoost Training Precision:', xgb_precision)
print('XGBoost Training Recall:', xgb_recall)
print('XGBoost Training F1:', xgb_f1)
```

# 2.	Visualize and interpret the results to identify the strengths and weaknesses of each algorithm.

```python
import numpy as np
import matplotlib.pyplot as plt

def compare_metrics(xgb_metrics, lgbm_metrics, title):
    """
    Compares the metrics of XGBoost and LightGBM using bar plots.

    Args:
        xgb_metrics: A dictionary of XGBoost metrics.
        lgbm_metrics: A dictionary of LightGBM metrics.
        title: The title of the plot.
    """

    metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
    xgb_values = [xgb_metrics['accuracy'], xgb_metrics['precision'], xgb_metrics['recall'], xgb_metrics['f1']]
    lgbm_values = [lgbm_metrics['accuracy'], lgbm_metrics['precision'], lgbm_metrics['recall'], lgbm_metrics['f1']]

    # Create the bar plot
    plt.figure(figsize=(10, 6))
    X = np.arange(4)
    plt.bar(X - 0.2, xgb_values, 0.4, label='XGBoost')
    plt.bar(X + 0.2, lgbm_values, 0.4, label='LightGBM')
    plt.xticks(X, metrics)
    plt.ylabel('Metric Score')
    plt.title(title)
    plt.legend(loc='best')
    plt.show()

# Create dictionaries to store metrics
xgb_metrics = {'accuracy': xgb_accuracy, 'precision': xgb_precision, 'recall': xgb_recall, 'f1': xgb_f1}
lgbm_metrics = {'accuracy': lgbm_accuracy, 'precision': lgbm_precision, 'recall': lgbm_recall, 'f1': lgbm_f1}

# Call the comparison function
compare_metrics(xgb_metrics, lgbm_metrics, 'XGBoost vs LightGBM Metrics')
```

- Summary of the Results:
1. Accuracy: Both XGBoost and LightGBM show similar or identical accuracy. This suggests that both models are equally effective at making correct predictions in terms of overall performance.
2. Precision: Precision is almost the same for both models, meaning both have a similar rate of correctly identifying true positives out of all positive predictions made.
3. Recall: Both models show similar recall, indicating that they are equally effective at identifying the relevant positive cases from the dataset.
4. F1-Score: The F1-Score, which balances precision and recall, is also very similar between XGBoost and LightGBM. This suggests that neither model has a strong advantage over the other in terms of handling the trade-off between precision and recall.
- Both models perform almost identically on all metrics. This could be due to the dataset's characteristics, where both gradient boosting algorithms (XGBoost and LightGBM) are highly optimized for similar tasks. In cases where the dataset is not very complex or has balanced classes, these models can perform similarly without much variance in results.
