# 2. Preprocess the data by handling missing values & Outliers, if any.

```python
df_concat.info()
```

```python
df_concat.isnull().sum()
```

```python
# handling outliers

# constructing Boxplot for Outlier Detection
plt.figure(figsize=(10, 6))
sns.boxplot(x = 'legs', data = df_concat, palette="Set3")
plt.title('Boxplot of Legs Feature')
plt.show()
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df_concat['legs'].quantile(0.25)
Q3 = df_concat['legs'].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Define the bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
outliers = df_concat[(df_concat['legs'] < lower_bound) | (df_concat['legs'] > upper_bound)]
print("Outliers in 'legs' feature:")
print(outliers)
```

```python
# Remove outliers (if necessary)
df_new = df_concat[~((df_concat['legs'] < lower_bound) | (df_concat['legs'] > upper_bound))]
```

```python
# Checking the new shape after removing outliers
print(f"New shape after outlier removal: {df_new.shape}")
```

- Outliers in the legs feature are not errors: The "outliers" like animals with 0 or many legs are natural and should be retained for the classification task. They are valuable for distinguishing animal types such as reptiles, insects, and mammals.
- Valid Outliers: In this specific case, the detected outliers (e.g., 0 legs for fish and snakes or many legs for insects) are valid data points that represent unique animal groups and contribute to the classification of different species.
-  Since the outliers in the legs feature reflect the natural biological differences between animal species, these should be kept in the dataset.

```python
training_accuracy_list = []
test_accuracy_list = []
```

```python
for k in range(5, 18, 2):
    training_accuracy = []
    test_accuracy = []

    for i in range(1, 100, 1):
        X_train, X_test, Y_train, Y_test = train_test_split(X_new,Y_new, test_size=0.30, random_state=i)
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train.values, Y_train)
        y_pred_train = knn.predict(X_train.values)
        y_pred_test = knn.predict(X_test.values)
        training_accuracy.append(accuracy_score(Y_train, y_pred_train))
        test_accuracy.append(accuracy_score(Y_test, y_pred_test))

    training_accuracy_list.append(np.mean(training_accuracy).round(2))
    test_accuracy_list.append(np.mean(test_accuracy).round(2))
```

```python
print("Training Accuracies:", training_accuracy_list)
print("Test Accuracies:", test_accuracy_list)
```

```python
import matplotlib.pyplot as plt
plt.scatter(range(5, 18, 2),training_accuracy_list,color='blue')
plt.plot(range(5, 18, 2),training_accuracy_list,color='black')
plt.scatter(range(5, 18, 2),test_accuracy_list,color='red')
plt.plot(range(5, 18, 2),test_accuracy_list,color='black')
plt.show()
```

```python
# Train the KNN model
knn_v = KNeighborsClassifier(n_neighbors=5)
knn_v.fit(X_train_v, Y_train_v)
```

```python
# CreatING a meshgrid for visualization
X_min, X_max = X_train_v[:, 0].min() - 1, X_train_v[:, 0].max() + 1
Y_min, Y_max = X_train_v[:, 1].min() - 1, X_train_v[:, 1].max() + 1
XX, YY = np.meshgrid(np.arange(X_min, X_max, 0.01), np.arange(Y_min, Y_max, 0.01))
```

```python
# Predicting class labels for the grid points
Z = knn_v.predict(np.c_[XX.ravel(), YY.ravel()])
Z = Z.reshape(XX.shape)
```

```python
# Plotting the decision boundary
plt.figure(figsize=(12, 8))
plt.contourf(XX, YY, Z, alpha=0.3, cmap='viridis')
plt.scatter(X_train_v[:, 0], X_train_v[:, 1], c=Y_train_v, edgecolor='k', marker='o', s=100)
plt.title('KNN Decision Boundaries with Legs and Hair Features')
plt.xlabel('Legs (scaled)')
plt.ylabel('Hair (scaled)')
plt.colorbar(label='Animal Type')
plt.show()
```
