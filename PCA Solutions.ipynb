# Task 2: Dimensionality Reduction with PCA:
1.	Standardize the features to ensure they have a mean of 0 and a standard deviation of Implement PCA to reduce the dimensionality of the dataset.
2.	Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.
3.	Transform the original dataset into the principal components.

```python
# Data Partition

X = df.drop(columns = 'Type')
X.head()
```

```python
Y = df['Type']
Y.head()
```

```python
# Standardization

from sklearn.preprocessing import StandardScaler
SS = StandardScaler()

df_scaling = SS.fit_transform(X)
```

```python
df_scaling
```

# Dimensionality Reduction with PCA 

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
pca.fit_transform(df_scaling)
```

```python
# explained variance by each component

exlplained_variance = pca.explained_variance_ratio_
exlplained_variance
```

```python
# Scree Plot

plt.figure(figsize = (10,6))
plt.plot(range(1, len(exlplained_variance) + 1), exlplained_variance, marker = 'o', linestyle = '--')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```

```python
# Cumulative Explained Variance

cumulative_variance = np.cumsum(exlplained_variance)
cumulative_variance
```

```python
# Cumulative Explained Variance plot

plt.figure(figsize = (10,6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker = 'o', linestyle = '--')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number ofPrincipal Component')
plt.ylabel('Cumulative Variance Explained')
plt.show()
```

```python
cumulative_variance
```

The cumulative explained variance plot shows that:

- About 80% of the variance is explained by the first 5 components.
- Approximately 90% is explained by the first 7 components.
- Scree Plot: This plot displays the explained variance of each principal component. In it, we often observe a "elbow" point where the explained variance begins to flatten.
-  Optimal Number of Components: Based on the scree plot, the best value for n_components would be around 6.
- This indicates that additional components contribute less to the overall variance.

```python
# Transforming the original dataset into the principal components(6)

pca = PCA(n_components = 6)
X_pca = pca.fit_transform(df_scaling)
X_pca
```

```python
pca.explained_variance_ratio_
```

```python
# Converting into DataFrame

pca_columns = [f'PC{i+1}' for i in range(X_pca.shape[1])]
df_pca = pd.DataFrame(X_pca, columns = pca_columns)
df_pca.head()
```

# Task 3: Clustering with Original Data:
1.	Apply a clustering algorithm (e.g., K-means) to the original dataset.
2.	Visualize the clustering results using appropriate plots.
3.	Evaluate the clustering performance using metrics such as silhouette score or DaviesBouldin index.

```python
from sklearn.cluster import KMeans
```

```python
inertia = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(df_scaling)
    inertia.append(kmeans.inertia_)
print(inertia)
```

```python
plt.figure(figsize = (11,4))
plt.plot(range(1,11), inertia)
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve')
plt.show()
```

```python
kcluster = KMeans(n_clusters=3,init='k-means++',random_state=42)
kcluster.fit(df_scaling)
```

```python
kcluster.labels_
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(df_scaling, labels = kcluster.labels_)
score.round(3)
```

# Task 5: Comparison and Analysis:
1.	Compare the clustering results obtained from the original dataset and PCA-transformed data.
2.	Discuss any similarities or differences observed in the clustering results.
3.	Reflect on the impact of dimensionality reduction on clustering performance.
4.	Analyze the trade-offs between using PCA and clustering directly on the original dataset.

1. Comparison of Clustering Results:

- Original Dataset: Clustering directly on the original dataset can leverage the full dimensionality, capturing all feature interactions. This can sometimes lead to highly granular clusters but may also introduce noise and outliers.
- PCA-Transformed Dataset: In the PCA-transformed dataset, clustering is based on a reduced number of principal components that explain most of the variance. This can lead to clusters that are more generalized but less susceptible to noise.

2. Similarity or Differences:

- Clusters in the PCA-transformed data may show improved compactness and separation, especially if the original dataset contained correlated features that were reduced by PCA.
- If your PCA retained a high amount of variance (e.g., 80-90%), clustering results should be reasonably similar to those on the original dataset.

3. Impact of Dimensionality Reduction on Clustering Performance:

- Computational Efficiency: Dimensionality reduction generally leads to faster clustering as the algorithm processes fewer features.
- Cluster Quality: In some cases, PCA can improve clustering quality by removing redundant or noisy information, leading to clearer, more interpretable clusters. However, reducing too much information can obscure clusters that depend on smaller, nuanced details.

4. Trade-Offs Between Using PCA and Clustering Directly:

- Interpretability: Clustering on the original dataset can be more interpretable since it uses the actual features. In contrast, clusters on PCA-transformed data might not directly correspond to original features, making interpretation harder.
- Efficiency vs. Completeness: PCA-based clustering is often more efficient and can yield meaningful clusters with fewer components. However, it may overlook minor but potentially relevant details that full-dimensional clustering captures.

1. Key Findings and Insights:

- Clustering Comparison: Clustering on both the original and PCA-transformed datasets yielded similar but distinct results. The PCA-transformed data led to more defined clusters, suggesting that dimensionality reduction can help by emphasizing key patterns in the data.
- Optimal Component Selection: The cumulative explained variance analysis demonstrated that around 57 principal components captured the majority of the datasets variability, balancing dimensionality reduction with information retention.

2. Practical Implications of Using PCA and Clustering:

- PCA in Data Analysis: PCA is especially useful in high-dimensional datasets where there is likely redundancy among features. By reducing dimensions, it highlights the most significant variance-driven features, helping analysts focus on the core structure of the data.
- Clustering: Clustering reveals inherent groupings within data, aiding in tasks like customer segmentation, anomaly detection, or classification in exploratory data analysis. The use of PCA before clustering can improve cluster quality and reduce noise in high-dimensional data.

3. Recommendations for Technique Usage:

- When to Use PCA: Use PCA when the dataset has a large number of correlated features or when computational efficiency is a concern. Its particularly beneficial when aiming to reduce noise and complexity without losing essential data structure.
- Direct Clustering on Original Data: Direct clustering is suitable when interpretability of clusters in terms of original features is important or if the dataset is already low-dimensional.
- Combined Approach (PCA + Clustering): For large, complex datasets, a combined PCA and clustering approach often provides better clustering quality while maintaining efficiency. This is ideal when working with high-dimensional data that may have underlying patterns or redundancy.
