# 1. Data Exploration and Preprocessing
●	Begin by loading and exploring the "Alphabets_data.csv" dataset. Summarize its key features such as the number of samples, features, and classes.

```python
import pandas as pd
df = pd.read_csv('Alphabets_data.csv')
df.shape
```

```python
# extracting Feature Names

df.columns.tolist()
```

```python
# Data Partition 
# Separating features and Target Variables

X = df.drop(columns = ['letter'])
Y = df[['letter']]
```

```python
X.head()
```

```python
Y.head()
```

```python
# Preprocessing : Data Normalization

from sklearn.preprocessing import MinMaxScaler
MM = MinMaxScaler()
```

```python
for i in X.columns:
    X[i] = MM.fit_transform(X[[i]])
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.20, random_state = 42)
```

```python
# One - Hot Encoding
y_train_encoded = pd.get_dummies(y_train['letter'], prefix = '', dtype = int)
y_test_encoded = pd.get_dummies(y_test['letter'], prefix = '', dtype = int)
```

```python
y_train_encoded.head()
```

# 3. Hyperparameter Tuning
●	Modify various hyperparameters, such as the number of hidden layers, neurons per hidden layer, activation functions, and learning rate, to observe their impact on model performance.

●	Adopt a structured approach like grid search or random search for hyperparameter tuning, documenting your methodology thoroughly.


```python
model = Sequential()
```

```python
model.add(Dense(128, activation = 'relu', input_shape = (X_train.shape[1],))) # input layer with 128 neurons
```

```python
model.add(Dense(64, activation = 'relu')) # hidden layer with 64 neurons
```

```python
model.add(Dense(26, activation = 'softmax')) # output layer with 26 neurons 
```

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
import numpy as np
```

```python
# Model Training

model.fit(X_train, y_train_encoded, epochs = 100, batch_size = 32, validation_data = (X_test, y_test_encoded) )
```

```python
# Model Evaluation

loss, accuracy = model.evaluate(X_test, y_test_encoded)
print('Test Loss:', loss)
print('Test Accuracy:', accuracy)
```

```python
# Making predictions on the Test Set

predictions = model.predict(X_test)
predicted_classes = [np.argmax(prediction) for prediction in predictions]
```

```python
# Decoding the predicted classes back to alphabet letters

alphabet_mapping = {i: chr(ord('A') + i) for i in range(26)}
predicted_letters = [alphabet_mapping[predicted_class] for predicted_class in predicted_classes]
predicted_letters
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
accuracy = accuracy_score(y_test, predicted_letters)
precision = precision_score(y_test, predicted_letters, average = 'macro')
recall = recall_score(y_test, predicted_letters, average = 'macro')
f1 = f1_score(y_test, predicted_letters, average = 'macro')
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
def create_model(neurons = 128, activation = 'relu', optimizer = 'adam'):
    model = Sequential()
    model.add(Dense(neurons, activation=activation, input_shape=(X_train.shape[1],)))
    model.add(Dense(neurons, activation=activation))
    model.add(Dense(26, activation='softmax'))
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
model = KerasClassifier(model = create_model, epochs = 10, batch_size = 32, verbose = 0, activation = 'relu', neurons = 64)
```

```python
# defining hyper parameter grid

param_grid = {
    'neurons' : [64,128,256],
    'optimizer' : ['adam','rmsprop','SGD']
}
```

```python
# GrisSearch with Cross-Validation

grid = GridSearchCV(estimator = model, param_grid = param_grid, cv = 5)
grid_result = grid.fit(X_train, y_train_encoded)
```

```python
best_model = grid_result.best_estimator_
best_params = grid_result.best_params_
print('Best Parameters:', best_params)
```

```python
predictions = best_model.predict(X_test)
predicted_classes = [np.argmax(prediction) for prediction in predictions]
```

```python
alphabet_mapping = {i: chr(ord('A') + i) for i in range(26)}  # Mapping for one-hot encoding to alphabet
predicted_letters = [alphabet_mapping[predicted_class] for predicted_class in predicted_classes]
```
