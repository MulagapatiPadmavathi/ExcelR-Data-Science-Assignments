# 4. Detect and treat outliers if necessary

```python
# Detect and treat outliers if necessary
# Detecting and Treating Outliers Using the Z-Score Method
# The Z-score method is another common technique, where any data point with a Z-score (standard deviation) greater than 3 or less than -3 is typically considered an outlier.

from scipy import stats
import numpy as np

# Calculate Z-scores for each value in the df_filled DataFrame
z_scores = np.abs(stats.zscore(df_filled))

# Filter the DataFrame to exclude rows where any Z-score is greater than 3
df_no_outliers_z = df_filled[(z_scores < 3).all(axis=1)]

# Print the shape of the DataFrame before and after outlier removal
print("Original dataset shape:", df_filled.shape)
print("Dataset shape after removing outliers (Z-score method):", df_no_outliers_z.shape)
```

# Since our dataset doesnâ€™t have any categorical variables,we can create categories from a numerical variable and then plot a bar chart or pie chart to display the frequency of these categories.
- Applying Binning to Multiple Variables
1. Define Bins: For each variable (e.g., LB and ASTV), we define bins that make sense based on the values in the dataset.
2. Convert to Categories: We use pd.cut() to group the numerical values into the categories defined by the bins.
3. Visualize: A bar chart and a pie chart are created for each of the new categorical-like variables.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Define bins and labels for each variable, ensuring number of labels is one less than number of bins
bin_definitions = {
    'LB': {
        'bins': [0, 110, 120, 130, 140, 160, 200],
        'labels': ['<110', '110-120', '121-130', '131-140', '141-160', '>160']
    },
    'AC': {
        'bins': [-1, 0, 2, 5, 10, 20],
        'labels': ['0', '1-2', '3-5', '6-10', '>10']
    },
    'FM': {
        'bins': [-1, 0, 5, 10, 20, 50],
        'labels': ['0', '1-5', '6-10', '11-20', '>20']
    },
    'UC': {
        'bins': [0, 0.5, 1.0, 2.0, 5.0, 10.0],
        'labels': ['<0.5', '0.5-1.0', '1.1-2.0', '2.1-5.0', '>5']
    },
    'DL': {
        'bins': [0, 1, 2, 5, 10, 15],
        'labels': ['0', '1', '2-5', '6-10', '>10']
    },
    'DS': {
        'bins': [0, 1, 2, 5, 10],
        'labels': ['0', '1', '2-5', '6-10', '>10']
    },
    'DP': {
        'bins': [0, 1, 2, 5, 10],
        'labels': ['0', '1', '2-5', '6-10', '>10']
    },
    'ASTV': {
        'bins': [0, 10, 20, 30, 40, 50, 100],
        'labels': ['<10', '10-20', '21-30', '31-40', '41-50', '>50']
    },
    'MSTV': {
        'bins': [0, 0.5, 1, 2, 5, 10],
        'labels': ['<0.5', '0.5-1.0', '1.1-2.0', '2.1-5.0', '>5']
    },
    'ALTV': {
        'bins': [0, 10, 20, 30, 40, 50, 100],
        'labels': ['<10', '10-20', '21-30', '31-40', '41-50', '>50']
    },
    'MLTV': {
        'bins': [0, 0.5, 1, 2, 5, 10],
        'labels': ['<0.5', '0.5-1.0', '1.1-2.0', '2.1-5.0', '>5']
    }
}

# Apply binning to all the defined variables
for col, bin_info in bin_definitions.items():
    try:
        # Check if bin labels match the bins correctly
        if len(bin_info['bins']) - 1 != len(bin_info['labels']):
            raise ValueError(f"Number of labels for {col} does not match number of bins minus one.")
        
        # Apply the binning process
        df_filled[f'{col}_Group'] = pd.cut(df_filled[col], bins=bin_info['bins'], labels=bin_info['labels'], right=False)
    except ValueError as e:
        print(f"Error in binning for column {col}: {e}")

# List of new categorical-like variables created from binning
grouped_vars = [f'{col}_Group' for col in bin_definitions.keys()]

# Create bar charts and pie charts for each grouped variable
for var in grouped_vars:
    if var in df_filled.columns:
        # Bar Chart
        plt.figure(figsize=(10, 6))
        df_filled[var].value_counts().plot(kind='bar')
        plt.xlabel(var)
        plt.ylabel('Count')
        plt.title(f'Bar Chart of {var}')
        plt.show()

        # Pie Chart
        plt.figure(figsize=(8, 8))
        df_filled[var].value_counts().plot(kind='pie', autopct='%1.1f%%')
        plt.title(f'Pie Chart of {var}')
        plt.ylabel('')
        plt.show()
```
