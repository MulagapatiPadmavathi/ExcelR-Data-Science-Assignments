# 4. Decision Tree Classification:

# Split the dataset into training and testing sets (e.g., using an 80-20 split).

```python
# Data Partition

Y = df['num']
X = df.drop('num', axis = 1)
```

```python
Y.head()
```

```python
# Data Splitting

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 42)
```

# Implement a Decision Tree Classification model using a library like scikit-learn.

```python
# Model Selection

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

# Train the model on the training set and evaluate its performance on the testing set using appropriate evaluation metrics (e.g., accuracy, precision, recall, F1-score, ROC-AUC).

```python
# Model Fitting

dt.fit(X_train, Y_train)
```

```python
# Predictions

y_pred_train = dt.predict(X_train)
y_pred_test = dt.predict(X_test)
```

```python
# Metrics

from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,roc_curve,roc_auc_score,auc
print('Training Accuracy :', accuracy_score(Y_train, y_pred_train))
print('Test Accuracy :', accuracy_score(Y_test, y_pred_test))
print('Precision Score(macro) :', precision_score(Y_test, y_pred_test, average = 'macro'))
print('Recall Score(macro) :', recall_score(Y_test, y_pred_test, average = 'macro'))
print('F1 Score(macro) :', f1_score(Y_test, y_pred_test, average = 'macro'))
print('Confusion Matrix :', confusion_matrix(Y_test, y_pred_test))
```

```python
cm = confusion_matrix(Y_test, y_pred_test)
TN = cm[0,0]
FP = cm[0,1]
TNR = TN/(TN + FP)
print('Confusion Matrix:\n', cm)
print("Specificity Score", TNR.round(2))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.preprocessing import label_binarize
y_test_bin = label_binarize(Y_test, classes=[0, 1, 2,3,4])
n_classes = y_test_bin.shape[1]
```

```python
y_prob = dt.predict_proba(X_test)
```

```python
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plotting all ROC curves for each class
plt.figure()
colors = ['blue', 'red', 'green']  
for i, color in enumerate(colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label='ROC curve for class {0} (AUC = {1:0.2f})'.format(i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--', lw=2)  
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) for Multiclass')
plt.legend(loc="lower right")
plt.show()
```

```python
for i in range(n_classes):
    roc_auc[i] = roc_auc_score(y_test_bin[:, i], y_prob[:, i])

# Printing AUC for each class
for i in range(n_classes):
    print(f"AUC for class {i}: {roc_auc[i]:0.2f}")
```

# 1. What are some common hyperparameters of decision tree models, and how do they affect the model's performance?

1. max_depth: Specifies the maximum depth of the tree.
- Effect on Performance:
1.A high value for max_depth allows the tree to grow deeper, which may result in a better fit to the training data. However, this can lead to overfitting, especially if the tree becomes too complex and captures noise.
2.A low value for max_depth leads to a simpler model, potentially underfitting the data, as the tree may not capture important patterns.
- Optimal Setting: A value that balances between underfitting and overfitting, often determined through cross-validation.

2. min_samples_split: The minimum number of samples required to split an internal node.
- Effect on Performance:
1.A high value for min_samples_split results in fewer splits and a simpler model (likely underfitting).
2.A low value allows the tree to grow more complex, making it more likely to overfit the training data by capturing too many details.
- Optimal Setting: A higher value helps avoid small splits, thus reducing overfitting, but too high a value may prevent the model from learning well.

3. min_samples_leaf: The minimum number of samples required to be at a leaf node.
- Effect on Performance:
1.A high value for min_samples_leaf forces the tree to have larger leaf nodes, which results in a simpler model with fewer branches and less risk of overfitting.
2.A low value allows the tree to create leaves with fewer samples, leading to more complexity and potential overfitting.
- Optimal Setting: A moderate value helps generalize better by ensuring leaf nodes don't become too small, which can capture noise.

4. criterion: The function used to measure the quality of a split. Common choices are: gini (Gini impurity),entropy (Information gain)
- Effect on Performance:
1.gini is faster to compute and is often used in practice. It works by minimizing the Gini impurity at each split.
2.entropy tends to provide slightly better results when there is a clear structure in the data, as it tries to maximize information gain at each split.
- Optimal Setting: Both usually perform similarly, but entropy can perform slightly better in some cases. It’s often useful to experiment with both.

5. max_features: The maximum number of features to consider when looking for the best split.
- Effect on Performance:
1.A low value means fewer features are considered, leading to less complex models with potentially less overfitting, but also possibly underfitting.
2.A high value allows the tree to consider more features, leading to more complex splits, and possibly overfitting.
- Optimal Setting: Typically sqrt (square root of the number of features) works well, especially for large datasets.

6. max_leaf_nodes: The maximum number of leaf nodes in the tree.
- Effect on Performance:
1.High value: The tree can become large and overfit.
2.Low value: Limits the number of splits and simplifies the model, potentially underfitting.
- Optimal Setting: A value that ensures the tree is not too large but still captures the underlying data structure.


7. class_weight: Weights associated with classes in the target variable.
- Effect on Performance:
1.Useful for handling imbalanced datasets by giving more weight to underrepresented classes.
2.None (default) gives equal weight to all classes.
- Optimal Setting: When working with imbalanced datasets, you can set class_weight='balanced' to adjust for the imbalance.

# 2. What is the difference between the Label encoding and One-hot encoding?

1. Label Encoding:
- Label Encoding transforms categorical labels into a numerical format, where each unique category is assigned a unique integer.

- How It Works: Each unique category is assigned a numerical value.
- Example: If you have a column with the categories ['red', 'green', 'blue'], label encoding will assign: red -> 0,green -> 1,blue -> 2
- When to Use:
- Ordinal Data: Label encoding is ideal for ordinal data where there is an inherent order or ranking between the categories (e.g., Low, Medium, High).
- Categorical Data with No Impact of Ordinality: It can also be used for nominal (non-ordinal) data but may not be suitable if the categorical variable doesn't have an order (it can lead to unintended ranking effects).

a. Advantages:
- Simple and memory efficient.
- Useful for models that can interpret numeric relationships, such as tree-based models like decision trees and random forests.

b. Disadvantages:
- Introduces Ordinal Relationships: For nominal variables (no inherent order), it may give the model the false impression that one category is greater than another (e.g., red = 0, green = 1, blue = 2 may imply an order that does not exist).
- Not suitable for linear models like logistic regression or neural networks, as they may interpret the encoded labels as continuous variables with a specific order.

2. One-Hot Encoding:
- One-Hot Encoding creates a binary (0 or 1) column for each category in the original categorical feature. For each record, only one column will have a value of 1 (indicating the category), and the others will be 0.

- How It Works:For a column with categories ['red', 'green', 'blue'], one-hot encoding creates three new columns:red -> [1, 0, 0],green -> [0, 1, 0],blue -> [0, 0, 1]
- When to Use:
- Nominal Data: One-hot encoding is ideal for nominal data where there is no order between the categories (e.g., colors, countries, etc.).

a. Advantages:
- No Ordinal Assumption: One-hot encoding treats each category equally, with no assumption of ordinality.
- Suitable for algorithms that are sensitive to the interpretation of values (e.g., linear regression, neural networks).

b. Disadvantages:
- Increased Dimensionality: One-hot encoding can significantly increase the number of features, especially for categorical variables with many unique values. For example, a column with 100 unique categories will generate 100 new features, which could lead to high computational cost and potential issues like the "curse of dimensionality."
- May result in sparse matrices when the number of categories is very large (mostly filled with zeros).
