```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression

# Define lists to store the results for each model
mse_list = []
rmse_list = []
r2_list = []
features_list = []

# Assuming df_combined is your dataset and df["Price"] is your target variable
Y = df["Price"]

# Define the feature names explicitly for clarity (replace with your actual column names)
feature_names = ["Age_08_04", "KM", "HP", "Automatic", "cc", "Doors", "Weight", "Diesel", "Petrol", "CNG", "Price"]
# Loop through each feature set, gradually adding features
for i in range(1, len(feature_names) + 1):
    # Dynamically create feature set X
    X = df_combined[feature_names[:i]]  # Select the first i features
    
    # Train the model
    LR = LinearRegression()
    LR.fit(X, Y)
    
    # Make predictions
    Y_pred = LR.predict(X)
    
    # Calculate metrics
    mse = mean_squared_error(Y, Y_pred).round(3)
    rmse = np.sqrt(mse).round(3)
    r2 = r2_score(Y, Y_pred).round(3)
    
    # Append results to lists
    mse_list.append(mse)
    rmse_list.append(rmse)
    r2_list.append(r2)
    
    # Store feature combination label
    features_list.append(" + ".join(feature_names[:i]))

# Create a DataFrame to store the results
results_df = pd.DataFrame({
    "Features": features_list,
    "MSE": mse_list,
    "RMSE": rmse_list,
    "R-squared": r2_list
})

# Remove duplicate rows (in case some models give the same results)
results_df = results_df.drop_duplicates()

# Print the results table
print(results_df)
```
